```python
ga4_table = "stg.ga4_weekly_stats"

date_dim_table = 'report.DateDim'
query = """
    SELECT DISTINCT dd.fiscalweek, dd.weekbegindate, dd.weekenddate
    FROM report.DateDim dd
    JOIN (
        SELECT fiscalyear, fiscalweek, fiscalweek - 1 AS previousFiscalWeek
        FROM report.DateDim
        WHERE DayDate = cast(getdate() as date)
    ) t1
    ON dd.fiscalweek = t1.previousfiscalweek AND dd.fiscalyear = t1.fiscalyear
"""
df_date = spark.read.format("jdbc")\
.option("driver", jdbcDriver)\
.option("url", jdbcUrl)\
.option("query",query)\
.option("user", jdbcUsername)\
.option("password", jdbcPassword)\
.load()
df_date.show()
gua_dates_range = df_date.collect()
fiscal_week = gua_dates_range[0][0]
week_begin_date = gua_dates_range[0][1]
week_end_date = gua_dates_range[0][2]
print("fiscal_week::",fiscal_week,"week_begin_date::",week_begin_date,"week_end_date::",week_end_date) 
display(df_date)
```

```python
property_id = 297479542
# property_brand  = {309713727:'Bates US'
# ,309561205:'CatFootwear Canada'
# ,309621431:'CatFootwear DE'
# ,309599233:'CatFootwear EMEAEmerging'
# ,309613069:'CatFootwear UK'
# ,309592771:'CatFootwear US'
# ,309589213:'Chacos US'
# #,309561793:'Grasshoppers US' -- GUA  not available 
# #,309607030:'Harley-DavidsonFootwear US' -- GUA  not available 
# ,309616895:'HushPuppies Canada'
# ,309650146:'HushPuppies US'
# ,309619975:'Hytest US'
# ,309607124:'Keds Canada'
# #,309609458:'Keds US' -- GUA  not available 
# ,309606225:'Merrell BE'
# ,309607810:'Merrell Canada'
# ,309615440:'Merrell DE'
# ,309628914:'Merrell EMEAEmerging'
# ,309620743:'Merrell ES'
# ,309579645:'Merrell FR'
# ,309621834:'Merrell NL'
# ,309591477:'Merrell SE'
# ,309599329:'Merrell UK'
# ,309592118:'OnlineShoes US'
# #,309626252:'Prokeds US' -- GUA  not available 
# ,310709711:'Saucony AT'
# ,309600562:'Saucony BE'
# ,309596198:'Saucony Canada'
# ,309624452:'Saucony DE'
# ,309643444:'Saucony EMEAEmerging'
# ,309629089:'Saucony ES'
# ,309607948:'Saucony FR'
# ,309537329:'Saucony IT'
# ,309603632:'Saucony NL'
# ,309587387:'Saucony UK'
# #,428511278:'Server-side GTM' -- GA4 and GUA  not available 
# ,309650574:'Sperry Canada'
# ,309632059:'Sperry US'
# ,309599656:'Wolverine Canada'
# ,309591384:'Wolverine US'
# ,297479542:'Saucony US'
# ,309617502:'Merrell US'}

property_brand  = {297479542:'Saucony US'}
date_range = DateRange(start_date="2024-05-26",end_date="2024-06-01")


    # Define the metrics and dimensions
metrics = [
        Metric(name="sessions"),
        Metric(name="addToCarts"),
        Metric(name="transactions"),
       # Metric(name="itemsPurchased"),
        Metric(name="purchaseRevenue"),   
        Metric(name="shippingAmount"),   
        Metric(name="taxAmount"),    
     
      #  Metric(name="transactionRevenue"), ## purchaseRevenue??
        #Metric(name="transactionShipping"),##shippingAmount
       # Metric(name="transactionTax") ## taxAmount
    ]
dimensions = [Dimension(name="deviceCategory")]

def initialize_analyticsreporting():
    """Initializes the Analytics Reporting API V4 service object.

    Returns:
      An authorized Analytics Reporting API V4 service object.
    """


    file_text = spark.read.text(f"{raw_adls_path}GA4/credentials.json", wholetext=True)
    credential_data = file_text.collect()[0][0]
  #print(json.loads(credential_data))
  #credentials = Credentials.from_authorized_user_info(json.loads(decoded_bytes))
    credentials = service_account.Credentials.from_service_account_info(json.loads(credential_data))
    client = BetaAnalyticsDataClient(credentials=credentials)
    return client

def get_report(client,p_metrics):
    """Queries the Google Analytics Data API.

    Args:
      analytics: An authorized Google Analytics Data API service object.
    Returns:
      The Google Analytics Data API response.
    """
    # Define the GA4 property ID
   

    # Define the date range for the data
    

    # Retrieve data from the Google Analytics 4 property
    #print("querying for metrics::",p_metrics)
    request1 = RunReportRequest(
        property=f"properties/{property_id}",
        date_ranges=[date_range],
        dimensions=dimensions,
        metrics=p_metrics,
          dimension_filter=FilterExpression(
                            filter=Filter(field_name="deviceCategory",
                            in_list_filter=Filter.InListFilter(
                                values=["tablet","desktop","mobile"]
                            ),
                            
                            )
                           ),
    )
    response =   client.run_report(request1) 
    return response

def extract_data(response):
  data = []
  for row in response.rows:
        #print(type(row))
        #dimension_values = [dimension.value for dimension in row.dimension_values]
        #metric_values = [metric.value for metric in row.metric_values]
        dimension_value = row.dimension_values[0].value
        session= int(row.metric_values[0].value)
        addToCarts= int(row.metric_values[1].value)
        transactions=int(row.metric_values[2].value)
        revenue = float(row.metric_values[3].value)
        shipping = float(row.metric_values[4].value)
        tax = float(row.metric_values[5].value)
        #print("Dimensions:", dimension_value)
        #print("Metrics:", session, ',',addToCarts,',',transactions)
        data.append((property_brand[property_id],dimension_value,session,addToCarts,transactions
        , revenue, shipping, tax))
  return data

def extract_qty(response):
    data = []
    for row in response.rows:
        #print(type(row))
        #dimension_values = [dimension.value for dimension in row.dimension_values]
        #metric_values = [metric.value for metric in row.metric_values]
        dimension_value = row.dimension_values[0].value
        quantity= int(row.metric_values[0].value)
        
        #print("Dimensions:", dimension_value)
        #print("Metrics:", session, ',',addToCarts,',',transactions)
        data.append((property_brand[property_id],dimension_value,quantity))
    return data    

# def print_response(response):
#     """Parses and prints the Google Analytics Data API response.

#     Args:
#       response: A Google Analytics Data API response.
#     """
#     print("response::",response)
#     for row in response.rows:
#         #print(type(row))
#         #dimension_values = [dimension.value for dimension in row.dimension_values]
#         #metric_values = [metric.value for metric in row.metric_values]
#         dimension_value = row.dimension_values[0].value
#         session= int(row.metric_values[0].value)
#         addToCarts= int(row.metric_values[1].value)
#         transactions=int(row.metric_values[2].value)
#        # print("Dimensions:", dimension_value)
#        # print("Metrics:", session, ',',addToCarts,',',transactions)

def main():
    client = initialize_analyticsreporting()

   
    response = get_report(client,metrics)
    print('completed getting report')
    data = extract_data(response)
    #print('extracted data::',data)
    df = spark.createDataFrame(data = data,schema= ga4_schema)
    df = df.withColumn('item_sales',round(lit(df.revenue-df.shipping-df.tax),2))
   
    qty_metrics =  [Metric(name="itemsPurchased")]
    #print_response(response)
    response = get_report(client,qty_metrics)
    print('completed getting report for itemsPurchased')
    #print("response for quantity::", response)
    qty_extraction = extract_qty(response)
    df_qty = spark.createDataFrame(data= qty_extraction, schema = ga4_qty_schema)
    #print('extract data for itemsPurchased::',qty_extraction)
    #df = df.withColumn('quantity',(lit(quantity)))
    df = df.groupBy('brand').pivot('device_category').sum('sessions','adds_to_cart','transactions','item_sales')
    df_qty= df_qty.groupBy('brand').pivot('device_category').sum('quantity')
    df_qty = df_qty.selectExpr('brand'
    ,"split(brand,' ')[1] as country" 
    ,"case when split(brand,' ')[1] = 'US' then 'US' when split(brand,' ')[1] = 'CA' then 'CA' else 'EU' end as region"
    ,'`desktop` as desktop_units'
    ,'`mobile` as phone_units'
    ,'`tablet` as tablet_units' )


    #df = df.groupBy('brand').pivot('device_category').sum('sessions','adds_to_cart','transactions')
    df = df.selectExpr('brand'
    ,"split(brand,' ')[1] as country" 
    ,"case when split(brand,' ')[1] = 'US' then 'US' when split(brand,' ')[1] = 'CA' then 'CA' else 'EU' end as region"
    ,'`desktop_sum(sessions)` as desktop_visits'
    ,'`mobile_sum(sessions)` as phone_visits'
    ,'`tablet_sum(sessions)` as tablet_visits'
    ,'`desktop_sum(adds_to_cart)` as desktop_carts'
    ,'`mobile_sum(adds_to_cart)` as phone_carts'
    ,'`tablet_sum(adds_to_cart)` as tablet_carts'
    ,'`desktop_sum(transactions)` as desktop_orders'
    ,'`mobile_sum(transactions)` as phone_orders'
    ,'`tablet_sum(transactions)` as tablet_orders'
    #  ,'`desktop_sum(quantity)` as desktop_units'
    #  ,'`mobile_sum(quantity)` as phone_units'
    #  ,'`tablet_sum(quantity)` as tablet_units'
     ,'`desktop_sum(item_sales)` as desktop_web_sales'
     ,'`mobile_sum(item_sales)` as phone_web_sales'
     ,'`tablet_sum(item_sales)` as tablet_web_sales'
    )
   
    #df = df.withColumn('country',split(df.brand,' ')[-1])
    display(df)
    display(df_qty)
    df_join = df.alias('d1').join(df_qty.alias('d2'),['brand','country','region'])
    df_join = df_join.withColumn('week_end',lit('2024-06-01'))
    df_join = df_join.selectExpr('brand','country'
                       ,'region','desktop_visits','phone_visits','tablet_visits'
                       ,'desktop_carts','phone_carts','tablet_carts'
                       ,'desktop_orders','phone_orders','tablet_orders'
                       ,'desktop_units','phone_units','tablet_units'
                       ,'desktop_web_sales','phone_web_sales','tablet_web_sales','week_end')
    print('writing to sql server...')                   
    df_join = df_join.write.format("jdbc")\
                    .option("driver", jdbcDriver)\
                    .option("url", jdbcUrl)\
                    .option("dbtable", ga4_table)\
                    .option("user", jdbcUsername)\
                    .option("password", jdbcPassword)\
                    .mode('append')\
                    .save()
    display(df_join)
main()
```
