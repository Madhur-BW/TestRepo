```python
#Recursive function to traverse path, move files, and delete empty folders
def deep_ls_move(path: str, dest: str):
    # List all files in path
    li = mssparkutils.fs.ls(path)

    # Output current path statistics
    print(len(li), path)

    # If empty, send back empty and delete
    if len(li) == 0:
        return 'empty'
    
    # Move all files
    for x in li:
        if x.size != 0:
            mssparkutils.fs.mv(src=x.path, dest=f'{dest}/{x.name}', create_path=True, overwrite=False) #Leaving overwrite as false in case there are duplicate filenames (I don't believe there will be)
        else:
            if deep_ls_move(path=x.path, dest=dest) == 'empty':
                mssparkutils.fs.rm(x.path)
```
